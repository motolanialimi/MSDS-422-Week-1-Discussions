```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data
titanicdata = pd.read_csv('train.csv',sep=',')
```

```python
# Show a sample of the first 10 rows of data
titanicdata.head(10)
```

```python
# Understand the data types of the data
titanicdata.info()
```

```python
# Summarize the numerical fields in the data
titanicdata.describe()
```

```python
# Exploratory data analysis
pd.plotting.scatter_matrix(titanicdata, diagonal='kde', color='black', alpha=0.3, figsize=(15,15))
```

```python
# Plot to show the Class and Frequency

plt.figure(figsize=(10,10))
plt.hist(titanicdata['Pclass'], 3, color='blue', alpha=0.3)
plt.xticks(titanicdata['Pclass'])
plt.title('Class Count')
plt.xlabel('Class')
plt.ylabel('Frequency')
```

```python
# Plot to show the Age and Frequency

plt.figure(figsize=(10,10))
titanicage = titanicdata['Age']
plt.hist(titanicage.dropna(), 20, color='red', alpha=0.3)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
# Plot to show the No of Sibling/Spouse Abroad and No of Parent/Child Abroad

plt.figure(figsize=(10,10))
plt.scatter(titanicdata['SibSp'], titanicdata['Parch'], color='purple', alpha=0.3)
plt.title('Scatterplot')
plt.xlabel('Number of Sibling/Spouse Aboard')
plt.ylabel('Number of Parent/Child Aboard')
```

```python
# Plot to show the Fare
plt.figure(figsize=(10,10))
plt.boxplot(titanicdata['Fare'], vert=False)
plt.title('Boxplot of Fare')
```
